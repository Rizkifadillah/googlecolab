# **Seaborn & Pandas Dataframe**

***Statistik Summery***

```python
df_path = '/content/drive/MyDrive/course_ipynb/Kumpulan_dataset/cereal.csv'
df = pd.read_csv(df_path)
df
```

***Missing Value dengan seaborn***

```python
plt.figure(figsize=(6,6))
sns.heatmap(df.isna(), cmap='Blues')
```

***Correlation Heatmap***

```python
plt.figure(figsize=(6,6))
sns.heatmap(df.corr(), cmap='bwr', cbar=False, vmin=-1, annot=True, square=True, fmt=".1f")
```

***Pairplot***

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue="shelf", vars=['calories', 'fat','rating'])
```

```python
sns.pairplot(df, hue="shelf", vars=['calories', 'fat','rating'], kind='reg') #dengan regresi
```

```python
# analisis 2 data
sns.jointplot("potass","fiber", data=df, marginal_kws={"bins": 25}, kind="reg")
```

# **Data bertipe kategori**

***Persebaran***

```python
sns.catplot("mfr", "rating", data=df)
```

```python
# boxplot
sns.boxplot("mfr", "rating", data=df)
```

```python
# violinplot
sns.violinplot("mfr", "rating", data=df)
```

***Count***

```python
sns.catplot("mfr", data=df, kind='count')
```

```python
sns.countplot("mfr", data=df)
```

```python
sns.countplot("mfr", data=df, hue='shelf')
```
