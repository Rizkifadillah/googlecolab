```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0, 10, 10)
y = np.sqrt(X)
y1 = np.sin(X)
y2 = np.cos(X)
```

# Initial Figure dan Axes

```python
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.plot(X,y)
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.plot(X,y1, 'g')
ax.plot(X,y2, 'r')
# atau
# plt.figure(figsize=(6,6))
# plt.plot(X,y1)
# plt.plot(X, y2)
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.plot(X,y1, 'g', linewidth=5.5)
ax.plot(X,y2, 'r', markersize=3)
ax.set(xlim=(0,10),title='sin dan cos', ylim=(-2,2))
ax.axis('equal')
# plt.xlim(0,10)
# plt.ylim(-2,2)
# plt.axis('equal')
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.plot(X,y1, 'g', linewidth=5.5)
ax.plot(X,y2, 'r', markersize=3)
ax.set(xlim=(0,10),title='sin dan cos', ylim=(-2,2), xlabel='x', ylabel='y')
ax.axis('equal')
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.plot(X,y1, 'g', linewidth=5.5)
ax.plot(X,y2, 'r', markersize=3)
ax.set_xlim(0,10)
ax.set_ylim(-2,2)
ax.set_title('Sin dan Cos')
ax.set_xlabel('x', fontsize=14, fontweight=500)
ax.set_ylabel('y', fontsize=14, fontweight=500)
ax.axis('equal')
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.plot(X,y1, 'g', linewidth=5.5)
ax.plot(X,y2, 'r', markersize=3)
ax.set_xlim(0,10)
ax.set_ylim(-2,2)
ax.set_title('Sin dan Cos')
ax.set_xlabel('x', fontsize=14, fontweight=500)
ax.set_ylabel('y', fontsize=14, fontweight=500)
ax.axis('equal');
```

# **3D Plot**

```python
from mpl_toolkits import mplot3d
```

```python
fig = plt.figure(figsize=(8,8)) # membuat figure
ax = plt.axes(projection='3d')  #membuat koordinat
```

```python
x = np.linspace(-6,6,5)
y = np.linspace(-6,6,5)
```

```python
X, Y = np.meshgrid(x,y)
```

```python
X
```

```python
Y
```

```python
Z = np.sqrt(X**2 + Y**2 +10)
```

```python
fig = plt.figure(figsize=(8,8)) 
ax = plt.axes(projection='3d')  
ax.plot_surface(X,Y,Z);
```

```python
x = np.linspace(-6,6,50)
y = np.linspace(-6,6,50)
X, Y = np.meshgrid(x,y)
Z = np.sqrt(X**2 + Y**2 +10)

fig = plt.figure(figsize=(8,8)) 
ax = plt.axes(projection='3d')  
ax.plot_surface(X,Y,Z);
```

```python
x = np.linspace(-6,6,50)
y = np.linspace(-6,6,50)
X, Y = np.meshgrid(x,y)
Z = np.sqrt(X**2 + Y**2 +10)

fig = plt.figure(figsize=(8,8)) 
ax = plt.axes(projection='3d')  
ax.plot_surface(X,Y,Z, antialiased=False, cmap='winter');
ax.set(xlabel='x', ylabel='y', zlabel='z')
```

```python
x = np.linspace(-6,6,500)
y = np.linspace(-6,6,500)
X, Y = np.meshgrid(x,y)
Z = np.sqrt(X**2 + Y**2 +10)

fig = plt.figure(figsize=(8,8)) 
ax = plt.axes(projection='3d')  
ax.plot_surface(X,Y,Z, antialiased=False, cmap='winter');
ax.set(xlabel='x', ylabel='y', zlabel='z')
ax.view_init(elev=-30, azim=45) #mengubah sudut pandang
```

```python
fig = plt.figure(figsize=(8,8)) 
ax = plt.axes(projection='3d')  
ax.plot_wireframe(X,Y,Z, antialiased=False, cmap='winter'); #wireframe
ax.set(xlabel='x', ylabel='y', zlabel='z')
ax.view_init(elev=-30, azim=45) #mengubah sudut pandang
```

```python
fig = plt.figure(figsize=(8,8)) 
ax = plt.axes(projection='3d')  
ax.contour3D(X,Y,Z, 50, cmap='winter'); #contour3D
ax.set(xlabel='x', ylabel='y', zlabel='z')
ax.view_init(elev=-30, azim=45) #mengubah sudut pandang
```
